# data cleaning 

```python
# delete unnamed column
df.drop('Unnamed: 0', axis=1, inplace=True)

df
```

```python
# Rename columns
new_column_names = {'f_title': 'persian_title', 'e_title': 'english_title', 'data_rating': 'rate', 'price_broken': 'price'
    , 'price_special': 'net_price', 'discount': 'discount_percent', 'publishers': 'publisher_name', 'author': 'writer',
    'book_id': 'id', 'shbook': 'ISBN', 'ghat': 'size', 'npages': 'num_page', 'translator': 'translator', 
        'jyear': 'solarh_py', 'myear': 'gregorian_py', 'typej': 'cover_type', 'sprint': 'print_series'}
df.rename(columns=new_column_names, inplace=True)
df = df.replace()

df
```

```python
# Clean publisher Name
# Remove square brackets from the specified column
df['publisher_name'] = df['publisher_name'].str.replace(r'\[|\]', '', regex=True)

# Print the DataFrame to see the result
df
```

```python
# Clean ISBN Column
# Remove non-digit characters from the specified column
df['ISBN'] = df['ISBN'].str.replace(r'\D', '', regex=True)

# Print the DataFrame to see the result
df
```

```python
# Clean NUm page 
# Extract and keep only digits in the specified column
df['num_page'] = pd.to_numeric(df['num_page'], errors='coerce')

# Print the DataFrame to see the result
df 
```

```python
# clean gregorian column
# Assuming 'your_column_name' is the name of the column you want to extract from
df['gregorian_py'] = df['gregorian_py'].str.extract(r'(\d{4})')


# Define a custom function to extract 4 digits or return NaN
def extract_four_digits(cell):
    if isinstance(cell, str) and cell.isdigit() and len(cell) == 4:
        return cell
    else:
        return np.nan

# Apply the custom function to the column and create a new column
df['gregorian_py'] = df['gregorian_py'].apply(extract_four_digits)

# # Remove single quotes from the specified column
# df['gregorian_py'] = df['geregorian_py'].str.replace("'", '')
# # Convert the column to datetime
# df['geregorian_py'] = pd.to_datetime(df['gregorian_py'])

# Print the DataFrame to see the result
df
```

```python
# Clean Solar-Hijri column
# Remove single quotes from the specified column
df['solarh_py'] = df['solarh_py'].str.replace("'", '')

# Print the DataFrame to see the result
df
```

```python
# Clean print series
# Extract and keep only digits in the specified column
df['print_series'] = pd.to_numeric(df['print_series'], errors='coerce')

# Print the DataFrame to see the result
df 
```

```python
file_path = "C:\\Pedram\\quera\\6th week\\Code\\DATABASE\\all_persons all.csv"
df_person = pd.read_csv(file_path)
df_person
```

```python
df_person.drop('Unnamed: 0', axis=1, inplace=True)

df_person
```

```python

# Define a function to check if the role is "writer"
def is_writer(role):
    return 1 if role == 'author' else 0

# Define a function to check if the role is "translator"
def is_translator(role):
    return 1 if role == 'translator' else 0

# Create new columns "is_writer" and "is_translator" based on the "role" column
df_person['is_writer'] = df_person['role'].apply(lambda x: is_writer(x))
df_person['is_translator'] = df_person['role'].apply(lambda x: is_translator(x))

# Print the DataFrame to see the result
df_person
```

```python
df_person.drop('book_id', axis=1, inplace=True)
df_person.drop('role', axis=1, inplace=True)
# Remove duplicates based on the 'name' column and select the largest value from each other column
df_result = df_person.groupby('name').agg(lambda x: x.max()).reset_index()

# # Remove duplicate rows based on the specified column
# 
# df_person_new = df_person.drop_duplicates(subset='name')
# df_person_new
df_result
```

```python
# Set the starting ID value
start_id = 10000

# Add an auto-incrementing ID column starting from 1000
df_result['person_id'] = range(start_id, start_id + len(df_result))


# Print the DataFrame to see the result
df_result
```

```python
# Create a session to insert data into the tables
Session = sessionmaker(bind=engine)
session = Session()

# Insert data from the DataFrame into person table
df_result[[ 'person_id','name','is_writer', 'is_translator']].to_sql('person', con=engine, if_exists='append', index=False)

# Commit changes and close the session
session.commit()
session.close()
```
